<div style = "background-color:lightblue; color:green">Exploratory Data Analysis(EDA) is the process of understanding the dataset before applying it for models<br>
or making any decision.<br>
It is mainly concerned with converting raw data into useful data by looking into the data, handaling missing values, detecting outliers and observing patterns/characteristics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import Series
from pandas import DataFrame

```

```python
data = pd.read_csv('/home/kartika/Downloads/Customer Churn.csv')
data.head()
```

```python
data[data['TotalCharges'] == ' ']
```

```python
data['TotalCharges'] = data['TotalCharges'].replace(" ",0)
data['TotalCharges'] = data['TotalCharges'].astype(float)
```

```python
data.info()
```

<div style = "color:blue; font-size:20px">Since, tenure is 0, no charges is recorded in total charges. So we repalced blank with 0 as dtype float.

<div style = "color:blue; font-size:20px">No duplicate data throughout the dataframe.

```python
data['SeniorCitizen'] = data['SeniorCitizen'].replace({0:'No',1:'Yes'})
#data['SeniorCitizen'] = data['SeniorCitizen'].map({0:'No',1:'Yes'})     Using Function Mappling

# def conv(x):
#     if x == 0:
#         return 'No'
#     else:
#         return 'Yes'
# data['SeniorCitizen'] = data['SeniorCitizen'].apply(conv)     Using Function Application

```

```python
ax = sns.countplot(x = 'Churn', data = data)
ax.bar_label(ax.containers[0])
plt.figure(figsize=(10,6))
plt.show()

```

```python
plt.figure(figsize = (3,4))

gb = data.groupby('Churn').agg({'Churn':'count'})

plt.pie(gb['Churn'], labels = gb.index, autopct = '%1.2f%%')
plt.title("Percentage of Customers by Churn")
plt.show()


# g = data['Churn'].groupby(data['Churn']).count()
# print(g)
# plt.pie(g, labels = g.index, autopct = '%1.2f%%')
# plt.show()
#Grouping in Series
```
